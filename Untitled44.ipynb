# <center>Handling Mising Data - fillna,interpolate, dropna</center>

```python
import pandas as pd
import numpy as np
df = pd.read_csv('weatherNC.csv')
df
```

```python
df.set_index('day', inplace=True)
df
```

# fillna

```python
newdf = df.fillna(0)
newdf
```

```python
newdf = df.fillna({'temperature': 0,'windspeed': 0,'event':'noevent'})
newdf
```

```python
newdf = df.fillna(method='ffill')
newdf
```

```python
newdf = df.fillna(method='bfill')
newdf
```

#interpolate 

```python
newdf = df.interpolate()
newdf
```

# dropna

```python
newdf = df.dropna()
newdf
```
